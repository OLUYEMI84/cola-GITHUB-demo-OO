## Task 1: Create a DataFrame

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from functools import reduce
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'San Francisco' , 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("\nData Types:\n", df.dtypes)
```

# Task 3: Handling Null Values

```python

df_nulls = pd.DataFrame({
    'A': [1, np.nan, 3],
    'B': [np.nan, 'hello', 'world'],
    'C': [np.nan, np.nan, np.nan]
})
print("\nDataFrame with Nulls:")
print(df_nulls)

# Fill NaNs
df_filled = df_nulls.fillna({
    'A': 0,
    'B': 'N/A',
    'C': 'Unknown'
})
print("\nDataFrame after filling Nulls:")
print(df_filled)
```

# Task 9: Iterators and Generators

```python

class EvenIterator:
    def __init__(self):
        self.num = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= 5:
            raise StopIteration
        self.num += 2
        self.count += 1
        return self.num

```

```python
def even_generator():
    num = 2
    for _ in range(5):
        yield num
        num += 2

print("\nEven Iterator:")
for num in EvenIterator():
    print(num, end=' ')

print("\nEven Generator:")
for num in even_generator():
    print(num, end=' ')
```

# Task 10: Map, Reduce, Filter

```python

from functools import reduce

numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
product = reduce(lambda x, y: x * y, numbers)
evens = list(filter(lambda x: x % 2 == 0, numbers))

print("\n\nMap (Squared):", squared)
print("Reduce (Product):", product)
print("Filter (Evens):", evens)
```

# Task 11: Object-Oriented Programming

```python

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r1 = Rectangle(5, 3)
r2 = Rectangle(7, 2)

print("\nRectangle 1 - Area:", r1.area(), ", Perimeter:", r1.perimeter())
print("Rectangle 2 - Area:", r2.area(), ", Perimeter:", r2.perimeter())
```

#Task 12: Pandas Data Analysis

```python

emp_data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'IT', 'Finance'],
    'Salary': [55000, 70000, 62000, 48000]
})

```

```python
# Average salary by department
avg_salary = emp_data.groupby('Department')['Salary'].mean()
print("\nAverage Salary by Department:")
print(avg_salary)
```

```python
# Employees with salary > 60000
high_earners = emp_data[emp_data['Salary'] > 60000]['Name']
print("\nEmployees with Salary > 60000:")
print(high_earners.tolist())
```

```python
# Add Bonus column (10%)
emp_data['Bonus'] = emp_data['Salary'] * 0.10
print("\nDataFrame with Bonus:")
print(emp_data)
```
