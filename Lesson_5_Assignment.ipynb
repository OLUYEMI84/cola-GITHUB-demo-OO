## Task 1: Load the Titanic dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 1: Load the Titanic dataset
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df = pd.read_csv(url)
```

## # Task 2: Inspect the data

```python
# Task 2: Inspect the data
print(df.head())
print(df.tail())
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

### EXPLANATION
This shows summary statistics (count, mean, standard deviation, min, max, and quartiles) for several features of Titanic passengers:

Survived: 38.6% survived (mean = 0.386, where 1 = survived, 0 = not).

Pclass: Most passengers were in lower classes (mean = 2.31; 1 = first, 3 = third).

Age: Average age was about 29.5 years, with a range from 0.42 to 80.

Siblings/Spouses Aboard: Most had none (median = 0), but some had up to 8.

Parents/Children Aboard: Again, most had none, with a few having up to 6.

Fare: Highly variable, with fares ranging from 0 to over 512 (mean ≈ 32).

```python
print(df.isnull().sum())
```

## # Task 3: Clean and preprocess the data

```python
# Create 'PclassCategory'
df['PclassCategory'] = df['Pclass'].map({1: 'First', 2: 'Second', 3: 'Third'})
df['PclassCategory']
```

```python
# Create 'CabinCategory' - mark as 'Unknown' if NaN, else 'Known'
df['CabinCategory'] = df['Pclass'].apply(lambda x: 'Known' if pd.notnull(x) else 'Unknown')
df['CabinCategory']
```

```python
# Handle missing values in 'Age' using mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Age'].isnull().sum()
```

```python
# Create a new feature 'Family'
df['Family'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']
df['Family']
```

## Task 4: Identify specific data points

```python
# Task 4: Identify specific data points (max fare)
highest_fare_passengers = df[df['Fare'] == df['Fare'].max()]


print("Highest fare passengers:\n", highest_fare_passengers)
```

### EXPLANATION

These are the three highest fare-paying passengers on the Titanic, each paying 512.33, all in first class:

All survived the disaster.

Two were men (ages 35 and 36), and one was a woman (age 35).

One (Cardeza) traveled with a parent/child; the others were alone.

All had known cabins, and only one had family aboard.

This highlights how first-class status and solo travel may have correlated with higher survival and luxury (as reflected in the fare).

```python
# Task 4: Identify specific data points (max fare)

largest_family_passengers = df[df['Family'] == df['Family'].max()]

print("Largest family passengers:\n", largest_family_passengers)
```

### EXPLANATION

These passengers are members of the largest family group on board — the Sage family, with 10 family members in total.

Key points:

All were in third class and paid a total fare of 69.55.

Ages range from 5 to 21 years old.

All had 8 siblings/spouses aboard and 2 parents/children aboard.

Sadly, none survived the disaster.

Their cabin was known, but being in third class likely limited their chances of escape.

This highlights how large third-class families faced extreme survival disadvantages on the Titanic.

## Task 6: Introduce multi-indexing

This structure is powerful for grouped analysis or multi-level filtering

## Task 7: Visualizations

```python
# Histogram of passenger ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### EXPLANATION

This histogram shows the distribution of passenger ages on the Titanic:

The most common age group was between 20 and 30 years old.

The distribution is right-skewed, meaning more younger passengers and fewer older ones.

Fewer passengers were over 60 years old, and very few were under 5 years.

Most Titanic passengers were young adults, with decreasing numbers as age increased.

```python
# Bar plot of survival rates by passenger class
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class.plot(kind='bar', figsize=(8, 5))
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()
```

### EXPLANATION

This bar chart shows the survival rate by passenger class on the Titanic:

1st Class passengers had the highest survival rate (over 60%).

2nd Class had a moderate survival rate (around 47%).

3rd Class had the lowest survival rate (about 25%).

There was a clear survival advantage for higher-class passengers, likely due to better cabin locations, earlier access to lifeboats, and prioritization during evacuation.

```python
# Heatmap of correlation
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### EXPLANATION

Survived:

Positively correlated with Fare (0.26): Higher fare, higher survival chance.

Negatively correlated with Pclass (−0.34): Lower class number (i.e., 1st class) linked to higher survival.

Pclass:

Strong negative correlation with Fare (−0.55): 1st class paid more.

Also negatively correlated with Age (−0.39): Older passengers tended to be in higher classes.

Family-related features:

Siblings/Spouses Aboard and Parents/Children Aboard have a strong positive correlation with Family (0.89 and 0.78 respectively), which likely represents total family size.

Moderate correlation (0.41) between SibSp and Parch: Families tended to travel with both spouse/siblings and children/parents.

```python
# Pair plot of selected features
sns.pairplot(df[['Age', 'Fare',  'Siblings/Spouses Aboard' , 'Parents/Children Aboard' , 'Survived']], hue='Survived')
plt.show()
```

### EXPLANATION

Age vs Survival: More survivors (orange) are seen in younger age groups, suggesting age influenced survival.

Fare vs Survival: Survivors generally paid higher fares.

Family Size (SibSp & Parch): Most passengers had 0-2 family members aboard. Survival is mixed, but having some family (not too many) seems linked to higher survival.

Diagonals: These show distributions for each variable. Survivors are more concentrated in lower ages and higher fares.

## Task 8: Analyze survival rates

```python

survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Survival rate by sex:\n", survival_by_sex)
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("Survival rate by class:\n", survival_by_class)
```

### EXPLANATION
The survival rates on the Titanic reveal clear patterns based on gender and passenger class:

By gender, women had a significantly higher chance of survival than men. About 74% of women survived, compared to only 19% of men. This reflects the “women and children first” policy followed during the evacuation.

By class, passengers in first class had the highest survival rate at around 63%, followed by second class at about 47%, and third class at just 24%. This suggests that access to lifeboats and rescue efforts may have favored those in higher classes.

In summary, being female and traveling in first class greatly increased the chances of surviving the Titanic disaster.

## Task 9: Age distribution

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age Distribution by Passenger Class and Sex')
plt.show()
```

### EXPLANATION
First class passengers (Pclass 1) tended to be older than those in 2nd or 3rd class, for both sexes.

Across all classes, females and males had similar age ranges, though males often had slightly higher medians and more outliers (especially in 1st class).

Third class passengers were generally younger, with a wider spread of ages and more outliers.

# Task 10: Fare distribution

```python
# Task 10: Fare distribution
plt.figure(figsize=(12, 6))
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df, split=True)
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()

print("EDA tasks completed.")
```

### EXPLANATION

First class passengers (Pclass 1) had the widest range of fares, with some very high values (outliers). Survivors tend to have paid slightly more on average.

Second and third class passengers had lower and more concentrated fare distributions, with smaller differences between survivors and non-survivors.

Survival chances appear positively associated with higher fare—especially in first class.
